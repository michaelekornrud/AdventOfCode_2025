```python
import numpy as np

with open('file.txt', 'r') as file:
    codes = file.read().strip().split('\n')

codes_np = np.array(codes)
print(codes_np[:10])
```

```python
results = []

for i, line in enumerate(codes_np):
    digits = [int(char) for char in line if char.isdigit()]
    
    if len(digits) >= 2:
        # Find the largest digit and its position
        max_digit = max(digits)
        first_digit_idx = digits.index(max_digit)
        
        # Check if largest digit is at the last position
        if first_digit_idx == len(digits) - 1:
            # Largest is at end, so it becomes second digit
            # Find second largest to be first digit
            remaining_digits = digits[:-1]  # All digits except the last one
            second_largest = max(remaining_digits)
            
            first_digit = second_largest
            second_digit = max_digit
            
            print(f"Largest digit {max_digit} is at end - becomes second digit")
        else:
            # Largest is not at end, find largest from remaining digits after it
            remaining_digits = digits[first_digit_idx + 1:]
            
            if remaining_digits:
                second_digit = max(remaining_digits)
                first_digit = max_digit
            else:
                # This shouldn't happen given the new logic, but keeping as fallback
                second_digit = max(digits[:first_digit_idx]) if first_digit_idx > 0 else 0
                first_digit = max_digit
        
        result = int(f"{first_digit}{second_digit}")
        
        print(f"Line {i}: {line}")
        print(f"  All digits: {digits}")
        print(f"  Largest digit: {max_digit} at position {first_digit_idx}")
        print(f"  First digit: {first_digit}, Second digit: {second_digit}")
        print(f"  Result: {result}")
        print()
        
        results.append(result)

print(f"All results: {results}")
print(f"Sum: {sum(results)}")
```

```python
def get_largest_12_digits(digits):
    """
    Get the 12 largest digits by removing digits to maximize the final number.
    Uses a stack-based approach to ensure we get the lexicographically largest result.
    """
    if len(digits) <= 12:
        return digits
    
    digits_to_remove = len(digits) - 12
    stack = []
    
    for digit in digits:
        # While we can still remove digits and the current digit is larger than the top of stack
        while stack and digits_to_remove > 0 and stack[-1] < digit:
            stack.pop()
            digits_to_remove -= 1
        stack.append(digit)
    
    # If we still need to remove digits, remove from the end
    while digits_to_remove > 0:
        stack.pop()
        digits_to_remove -= 1
    
    return stack[:12]  # Ensure we return exactly 12 digits

# Test with the specific example
test = "234234234234278"
digits = [int(char) for char in test if char.isdigit()]
print(f"Original: {digits}")

result = get_largest_12_digits(digits)
number = int(''.join(map(str, result)))
print(f"Result: {result}  {number}")
print("Expected: 434234234278")

# Test with other examples
test_cases = [
    "987654321111111",  # Should be 987654321111
    "811111111111119",  # Should be 811111111119
    "234234234234278",  # Should be 434234234278
    "818181911112111"   # Should be 888911112111
]

print("\nTesting all examples:")
for test in test_cases:
    digits = [int(char) for char in test if char.isdigit()]
    result = get_largest_12_digits(digits)
    number = int(''.join(map(str, result)))
    print(f"{test}  {number}")
```

```python
results_2 = []
for index, line in enumerate(codes_np):
    digits = [int(char) for char in line if char.isdigit()]
    result = get_largest_12_digits(digits)
    number = int(''.join(map(str,result)))
    
    print(f"Line {index} : {line}")
    print(f"  All digits: {digits}")
    print(f"  Number : {number}" )

    results_2.append(number)

print(f"Result: {results_2}")
print(f"Sum of results: {sum(results_2)}")
```
