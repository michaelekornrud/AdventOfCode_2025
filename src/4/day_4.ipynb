```python
import numpy as np

with open('file.txt', 'r') as file:
    codes = file.read().strip().split('\n')


for index, code in enumerate(codes):
    print(f"Index: {index}, {code}")
```

```python
directions = [
    (-1, -1),  # Up-Left
    (-1,  0),  # Up
    (-1,  1),  # Up-Right
    ( 0, -1),  # Left
    ( 0,  1),  # Right
    ( 1, -1),  # Down-Left
    ( 1,  0),  # Down
    ( 1,  1)   # Down-Right
]

grid = [list(row) for row in codes]

result_grid = [row[:] for row in grid]

for row in range(len(grid)):
    for col in range(len(grid[row])):
        if grid[row][col] == '@':
            neighbor_count = 0

            for dr, dc in directions:
                neighbor_row = row + dr
                neighbor_col = col + dc

                if (0 <= neighbor_row < len(grid) and 
                    0 <= neighbor_col < len(grid[neighbor_row]) and 
                    grid[neighbor_row][neighbor_col] == '@'):

                    neighbor_count += 1

            if neighbor_count < 4 :
                result_grid[row][col] = 'x'

count_rols = sum(row.count('x') for row in result_grid)
print(f"Remaining 'x' : {count_rols}")

for i, row in enumerate(result_grid[:5]):
    print(f"Row {i}: {''.join(row)}")

            
```

```python
total_rolls = 0
iteration = 0

grid = [list(row) for row in codes]


while True:
    iteration += 1
    result_grid = [row[:] for row in grid]
    removed_this_round = 0

    for row in range(len(grid)):
        for col in range(len(grid[row])):
            if grid[row][col] == '@':
                neighbor_count = 0

                for dr, dc in directions:
                    neighbor_row = row + dr
                    neighbor_col = col + dc

                    if (0 <= neighbor_row < len(grid) and 
                        0 <= neighbor_col < len(grid[neighbor_row]) and 
                        grid[neighbor_row][neighbor_col] == '@'):

                        neighbor_count += 1

                if neighbor_count < 4 :
                    result_grid[row][col] = '.'
                    removed_this_round += 1

    print(f"Removed this round : {removed_this_round}")
    total_rolls += removed_this_round

    if removed_this_round == 0:
        break

    grid = [row[:] for row in result_grid]

# Count remaining '@' symbols
remaining_rolls = sum(row.count('@') for row in grid)
print("\nFinal results:")
print(f"Total removed: {total_rolls}")
print(f"Remaining rolls: {remaining_rolls}")

# Show final grid state
for i, row in enumerate(grid):  # Show first 10 rows
    print(f"Row {i}: {''.join(row)}")
```
