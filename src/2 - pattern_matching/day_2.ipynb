```python
import numpy as np 

with open('file.txt', mode='r') as file: 
    codes = file.read().strip().split(',')

codes_np = np.array(codes)
print("As NumPy array:", codes_np[:10])

```

```python
import builtins

ranges = []
invalid_ranges = []
sum_of_invalids = 0
for code in codes_np:
    ranges.append(code.split('-'))

for r in ranges:
    start = r[0].strip()
    end = r[1].strip()

    for num in builtins.range(int(start), int(end) + 1):
        # convert the integer to string before taking its length / slicing
        num_str = str(num)

        num_str = num_str.lstrip('0')

        mid = len(num_str) // 2
        first_half = num_str[:mid]
        second_half = num_str[mid:]

        if first_half == second_half:
            invalid_ranges.append(num_str)
        
print(invalid_ranges)

for i in invalid_ranges:
    sum_of_invalids += int(i)


print(f"Sum of invalid id's: {sum_of_invalids}")
    
```

```python
def is_invalid_id(id_str):
    n = len(id_str)
    
    for pattern_length in builtins.range(1, n // 2 + 1):
        pattern = id_str[:pattern_length]
        
        if pattern * (n // pattern_length) == id_str:
            # Make sure it repeats at least twice
            if n // pattern_length >= 2:
                return True, pattern, n // pattern_length
    
    return False, None, 0

ranges = []
invalid_ranges = []
sum_of_invalids = 0
for code in codes_np:
    ranges.append(code.split('-'))

for r in ranges:
    start = r[0].strip()
    end = r[1].strip()

    for num in builtins.range(int(start), int(end) + 1):
        # convert the integer to string before taking its length / slicing
        num_str = str(num)

        num_str = num_str.lstrip('0')

        is_invalid, pattern, repeat_count = is_invalid_id(num_str)

        if is_invalid:
            print(f"{num_str}: invalid ('{pattern}' repeated {repeat_count} times)")
            invalid_ranges.append(num_str)
        
print(f"\nFound {len(invalid_ranges)} invalid IDs")
print(f"Invalid IDs: {invalid_ranges}")

for i in invalid_ranges:
    sum_of_invalids += int(i)


print(f"\nSum of invalid id's: {sum_of_invalids}")
```
