```python
import numpy as np

with open('file.txt', 'r') as file:
    codes = file.read().strip().split('\n')

```

```python
ranges = []
ingredients = []

for line in codes:
    if '-' in line:
        ranges.append(line)
    else:
        ingredients.append(line)
    

good_food = set()
count = 0

for ingredient in ingredients:
    if ingredient and ingredient.strip():  # Check if not empty
        ingredient_num = int(ingredient)
        
        # Check if this ingredient falls in ANY range
        is_good = False
        for r in ranges:
            start, end = r.split('-')
            start, end = int(start), int(end)
            
            if start <= ingredient_num <= end:
                is_good = True
                break  # Found a match, no need to check other ranges
        
        if is_good:
            good_food.add(ingredient_num)
            count += 1

print(f"Total good ingredients: {count}")
```

```python
ranges = []

for line in codes: 
    if '-' in line: 
        start, end = line.split('-')
        ranges.append((int(start), int(end)))

ranges.sort()

merged_ranges = []

for start, end in ranges:
    if merged_ranges and start <= merged_ranges[-1][1] + 1:
        # Overlapping or adjacent - merge with prev. eange
        merged_ranges[-1] = (merged_ranges[-1][0], max(merged_ranges[-1][1], end))
    else:
        # No overlap - add new range
        merged_ranges.append((start,end))

# Calculate total count without generating individual numbers
total_fresh_ids = 0
for start, end in merged_ranges:
    total_fresh_ids += (end - start + 1)

print(f"Total fresh ingredient IDs: {total_fresh_ids}")
print(f"Number of merged ranges: {len(merged_ranges)}")

```
