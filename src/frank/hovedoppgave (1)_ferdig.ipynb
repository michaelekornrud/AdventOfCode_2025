L칮sningen demonstrerer avansert datavisualisering og statistisk analyse med multiple diagramtyper.

```python
# Ekstra utfordring - Flere analyser:

# 1. Gjennomsnittlig vibrasjon per dag
def beregn_gjennomsnitt_per_dag(ukedata):
    gjennomsnitt_per_dag = []
    for dag in ukedata:
        gjennomsnitt = sum(dag) / len(dag)
        gjennomsnitt_per_dag.append(gjennomsnitt)
    return gjennomsnitt_per_dag

gjennomsnitt_per_dag = beregn_gjennomsnitt_per_dag(ukedata)

# 2. Visualisering: Sammenligning av maks, gjennomsnitt og min per dag
min_per_dag = [min(dag) for dag in ukedata]

plt.figure(figsize=(12, 6))
dag_labels = [f"Dag {i+1}" for i in range(len(ukedata))]

x_pos = range(len(dag_labels))
width = 0.25

plt.bar([x - width for x in x_pos], maks_per_dag, width, label='Maks', alpha=0.8)
plt.bar(x_pos, gjennomsnitt_per_dag, width, label='Gjennomsnitt', alpha=0.8)
plt.bar([x + width for x in x_pos], min_per_dag, width, label='Min', alpha=0.8)

plt.title('Vibrasjonsm친linger - Sammenligning per dag')
plt.xlabel('Dag')
plt.ylabel('Vibrasjon (g)')
plt.xticks(x_pos, dag_labels)
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# 3. Statistikk utskrift
print("Statistikk for uken:")
print("-" * 40)
for i, (maks, avg, minimum) in enumerate(zip(maks_per_dag, gjennomsnitt_per_dag, min_per_dag)):
    print(f"Dag {i+1}: Maks={maks:.3f}g, Avg={avg:.3f}g, Min={minimum:.3f}g")

print("\nUke-statistikk:")
print(f"H칮yeste m친ling: {max(maks_per_dag):.3f}g")
print(f"Laveste m친ling: {min(min_per_dag):.3f}g")
print(f"Gjennomsnitt for uken: {sum(gjennomsnitt_per_dag)/len(gjennomsnitt_per_dag):.3f}g")

# 4. Identifiser dager med h칮y aktivitet (over 0.25g)
print("\nDager med h칮y vibrasjon (>0.25g):")
for i, maks in enumerate(maks_per_dag):
    if maks > 0.25:
        print(f"Dag {i+1}: {maks:.3f}g")
```
