```python
oppg_d_tall = [4, 7, 12, 15, 18, 21, 22]

antall_partall = 0
antall_oddetall = 0

# Skriv din kode her:

for tall in oppg_d_tall: 
    if tall % 2 == 0:
        antall_partall += 1

    else:
        antall_oddetall += 1

print(f"Antall partall: {antall_partall}")
print(f"Antall oddetall: {antall_oddetall}")
```

### Oppgave G

Du f친r en liste som inneholder flere lister (en `matrise`).
Skriv ut summen av hver rad ved 친 bruke en l칮kke inni en annen.

游눠 Hint: Bruk en ytre for-l칮kke som g친r gjennom radene, og en indre for-l칮kke som legger sammen tallene i raden (skriv en sum per linje).

```python
matrise = [
    [1, 2, 10],
    [8, -2, 0],
    [14, 4, 4]
]

# Skriv din kode her:
summer = []     # Setter opp en tom liste

for rad in matrise: # Ytre for-l칮kke
    sum = 0 # Sum settes til 0 for hver rad
    for nummer in rad: # Indre for-l칮kke
        sum += nummer  # Legger tallet til i summen
    
    summer.append(sum)  # Legger summen for gjeldende rad inn i listen. 

# Enumerate setter indeks p친 elementene i listen. 
for indeks, sum in enumerate(summer): # Bruker for-l칮kke for 친 hente ut hver enkelt element i listen. 
    print(f"Sum rad {indeks + 1} = {sum}")
```
