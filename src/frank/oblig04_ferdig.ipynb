### Oppgave F

I denne oppgaven skal vi se litt p친 "vasking" eller filtrering av data.  
Skriv en funksjon som tar inn en liste med blandede verdier og returnerer en ny liste som bare inneholder heltall.

- Lag et beskrivende funksjonsnavn i snake_case.
- Funksjonen tar en parameter (liste).
- G친 igjennom lista med en l칮kke.
- Behold kun elementer som er heltall (int).
- Returner den nye lista (ikke print inne i funksjonen).
- Test funksjonen med listen kilde_data gitt under.

游눠 Hint: 
- Vi kan sjekke p친 type med, `if type(element) is int:`
- Funksjonen `.append()` legger til et element i en liste.

````python
kilde_data = [1, "2", 3.5, -7, None, "hei", 0, 4.0, True]

# Skriv din kode her:
def sjekk_verdier (liste):
    ny_liste = []
    for element in liste:
        if type(element) is int:
            ny_liste.append(element)
    
    return ny_liste

liste = sjekk_verdier(kilde_data)

print("Ny liste:", liste)



````

### Oppgave G

Lag en enkel kalkulator.

Skriv en funksjon kalkuler(x, y, operasjon) der de to f칮rste parametere er tall (int) og operasjon er en tekst (string)  
F.eks: "add", "subtract", "multiply" eller "divide".

- Funksjonen skal returnere riktig resultat basert p친 operasjonen.
- Hvis operasjon ikke gjenkjennes, returner "Unknown operation".
- Test funksjonen med forskjellige operasjoner.
- (Utfordring) Fors칮k 친 bruke kun et return i funksjonen.
- (Utfordring) H친ndter deling p친 null ved 친 returnere f.eks. "Division by zero".

游눠En god m친te 친 kun ha en return er 친 lage en mellomvariabel.  

````python
# Skriv din kode her

def kalkuler(x, y, operasjon):
    sum = 0 
    if operasjon == "add":
        sum = x + y
    elif operasjon == "subtract":
        sum = x - y 
    elif operasjon == "multiply":
        sum = x * y
    elif operasjon == "divide":
        if y != 0 :
            sum = x / y 
        else: 
            sum = "Division by zero"
    else:
        sum = "Operasjon ikke kjent"
    return sum 


print("Summen av 2 + 3 er:", kalkuler(2,3, "add"))
print("Summen av 2 - 3 er:", kalkuler(2,3,"subtract"))
print("Summen av 5 * 5 er:", kalkuler(5,5,"multiply"))
print("Summen av 100 / 10 er:", kalkuler(23210, 10, "divide"))
print("Summen av 100 / 0 er:", kalkuler(100, 0, "divide"))

````
